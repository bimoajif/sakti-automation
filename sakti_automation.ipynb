### 2. Masuk ke Aset Tetap > RUH > Transaksi BMN > Usulan Penghapusan BMN > Pencatatan Barang Yang Akan Dihapuskan

```python
def hapus_bmn():
    wait = WebDriverWait(driver, 20)
    overlay = (By.CSS_SELECTOR, "div.ui-table-loading.ui-widget-overlay")
    # Masuk ke halaman Pencatatan Barang Yang Akan Dihapus
    elem = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[1]/p-scrollpanel/div/div[1]/div/div/app-menu/ul/li[5]/ul/li[1]/ul/li[1]/ul/li[5]/ul/li[1]/a")
    elem.click()

    # Klik Tombol Rekam
    element = wait.until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-container/app-aset-main-grid/div/p-panel/div/div[2]/div/div[4]/div/button[1]"))
        )

    time.sleep(2)
    element.click()

    # Klik Tombol Titik Tiga (...)
    element = wait.until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[1]/p-fieldset/fieldset/div/div/div[1]/div[3]/button"))
        )

    time.sleep(2)
    element.click()
    try:
        # Tunggu sampai baris tabel muncul
        rows = wait.until(
            EC.presence_of_all_elements_located(
                (By.XPATH, "//tbody[contains(@class,'ui-table-tbody')]/tr")
            )
        )

        i = 0

        # for i in range(len(rows)):
        while i < len(rows):
            # Re-locate rows after every click (Angular refresh)
            rows = driver.find_elements(By.XPATH, "//tbody[contains(@class,'ui-table-tbody')]/tr")
            cell = rows[i].find_element(By.TAG_NAME, "td")
            cells = rows[i].find_elements(By.TAG_NAME, "td")

            texts = [cell.text for cell in cells]

            is_exist = len(df[(df['Kode Barang'] == texts[0]) & (df['NUP'] == texts[2])]) != 0

            if(is_exist):
                process_cell(wait, overlay, cell, texts)
                i += 1            
            else:
                i += 1 

        elem = wait.until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/p-dynamicdialog/div[2]/div[2]/app-ast-cari-barang-asset-form/div[3]/div/button[2]"))
        )
        elem.click()
    finally:
        webdriver.ActionChains(driver).send_keys(Keys.ESCAPE).perform()


def process_cell(wait: WebDriverWait, overlay: tuple, cell, texts):
     # ðŸ”‘ WAIT UNTIL OVERLAY IS GONE

    print(texts)
    wait.until(EC.invisibility_of_element_located(overlay))

    driver.execute_script(
        "arguments[0].scrollIntoView({block:'center'});",
        cell
    )

    # Use JS click to bypass interception
    driver.execute_script("arguments[0].click();", cell)

    time.sleep(1)

    pilih = driver.find_element(By.XPATH, "/html/body/p-dynamicdialog/div[2]/div[2]/app-ast-cari-barang-asset-form/div[3]/div/button[1]")
    pilih.click()

    tgl_pembukuan = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[1]/p-fieldset/fieldset/div/div/div[5]/div[2]/p-calendar/span/input")
    tgl_pembukuan.clear()
    tgl_pembukuan.send_keys("3/12/2025")

    # TODO : klik tombol tgl pembukuan
    close_tgl_pembukuan = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[2]/p-fieldset[2]/fieldset/div/div/div[1]/div[2]/input")
    close_tgl_pembukuan.click()

    no_sk = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[2]/p-fieldset[2]/fieldset/div/div/div[1]/div[2]/input")
    no_sk.clear()
    no_sk.send_keys("KET-105/KPP.3302/2025")

    tgl_sk = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[2]/p-fieldset[2]/fieldset/div/div/div[2]/div[2]/p-calendar/span/input")
    tgl_sk.clear()
    tgl_sk.send_keys("3/11/2025")

    # TODO : klik tombol calendar tgl sk
    close_tgl_sk = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[2]/p-fieldset[2]/fieldset/div/div/div[2]/div[2]/p-calendar/span/input")
    close_tgl_sk.click()

    ket = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[2]/p-fieldset[2]/fieldset/div/div/div[3]/div[2]/input")
    ket.clear()
    ket.send_keys("Penghentian Penggunaan BMN")

    # simpan = driver.find_element(By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[4]/div[3]/button[1]")
    # simpan.click()

    time.sleep(2)

     # Klik Tombol Titik Tiga (...)
    element = wait.until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/app-root/app-main-layout/div/div[2]/app-ast-pencatatan-usulan-hapus-form/div/p-panel/div/div[2]/div/div[3]/div[1]/p-fieldset/fieldset/div/div/div[1]/div[3]/button"))
        )

    time.sleep(2)
    element.click()
```

```python
for _ in range(5):
    hapus_bmn()
```
